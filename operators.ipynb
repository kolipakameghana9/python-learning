OPERATORS


```python
#Arithmetic Operations
a=10
b=5
print(a+b) #Addition
print(a-b) #subtraction
print(a*b) #multiplication
print(a/b) #division
print(a//b) #floor division
print(a%b) #modulus
print(a**b) #exponentiation
```

```python
#comparison operators
a=10
b=5
print(a == b)  # Equal to
print(a != b)  # Not equal to
print(a > b)   # Greater than
print(a < b)   # Less than
print(a >= b)  # Greater than or equal to
print(a <= b)  # Less than or equal to
```

```python
#Assignment Operators
a=10
a+=5 # a = a + 5
print(a)
a-=3 # a = a - 3
print(a)
a*=2 # a = a * 2
print(a)
a/=2 # a = a / 2
print(a)
a//=2 # a = a // 2
print(a)
a%=3 # a = a % 3
print(a)

```

```python
#Logical Operators
age = 20
has_id = True

print(age >= 18 and has_id)
```

```python
age = 16
has_id = True

print(age >= 18 and has_id)
```

```python
age = 16
has_id = True

print(age >= 18 or has_id)
```

```python
is_raining = False

print(not is_raining)
```

```python
a = True
b = False

print(a and b)
print(a or b)
print(not a)
```

```python
#membership Operators
x = [1, 2, 3, 4, 5]
print(3 in x)
print(6 in x)
```
