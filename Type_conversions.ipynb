Type Conversion

#exmaple 
string → int
int → float
float → int

```python
#implicit conversion
x=2
y=3.5
z=x+y
print(z)
print(type(z))
```

```python
#explicit conversion
a="5"
b=int(a)
print(b)
print(type(b))
```

```python
a=10
b=float(a)
print(b)
print(type(b))
```

```python
a=7.5
b=int(a)
print(b)
print(type(b))
```

```python
a=100
b=str(a)
print(b)    
print(type(b))
```
